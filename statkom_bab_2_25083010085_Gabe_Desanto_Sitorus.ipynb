```python
#Bagian A UJI NORMALITAS JARQUE BERA
data = [12, 15, 14, 10, 18, 16, 13, 17, 11, 14]
```

Menyiapkan Data

Pada tahap ini data yang akan diuji normalitasnya disimpan ke dalam sebuah list bernama data. Setiap elemen dalam list merepresentasikan satu observasi atau satu nilai pengamatan. Data inilah yang akan digunakan dalam seluruh proses perhitungan Uji Jarque-Bera.

```python
#FUNGSI UJI NORMALITAS JARQUE BERA BAGIAN A NOMOR 1
# Langkah 1: Menghitung jumlah data
n = len(data)

# Langkah 2: Menghitung rata-rata
mean = sum(data) / n

# Langkah 3: Menghitung selisih atau deviasi setiap data terhadap rata-rata
selisih = []
for x in data:
    selisih.append(x - mean)

# Langkah 4: Menghitung momen ke-2 (m2)
m2 = 0
for d in selisih:
    m2 += d**2
m2 = m2 / n

# Langkah 5: Menghitung momen ke-3 (m3)
m3 = 0
for d in selisih:
    m3 += d**3
m3 = m3 / n

# Langkah 6: Menghitung momen ke-4 (m4)
m4 = 0
for d in selisih:
    m4 += d**4
m4 = m4 / n

# Langkah 7: Menghitung Kurtosis
kurtosis = m4 / (m2 ** 2)

# Langkah 8: Menghitung Skewness
skewness = m3 / (m2 ** 1.5)

# Langkah 9: Menghitung Statistik Jarque-Bera
jb = (n / 6) * (skewness**2 + ((kurtosis - 3)**2) / 4)

# Langkah 10: Menentukan nilai kritis Chi-Square
chi_square_critical = 5.991

# Langkah 11: Pengambilan keputusan
if jb > chi_square_critical:
    keputusan = "Tolak H0: Data tidak berdistribusi normal"
else:
    keputusan = "Gagal Menolak H0: Data berdistribusi normal"

# Langkah 12: Menampilkan hasil
print("n =", n)
print("mean =", mean)
print("selisih =", selisih)
print("m2 =", m2)
print("m3 =", m3)
print("m4 =", m4)
print("kurtosis =", kurtosis)
print("skewness =", skewness)
print("JB =", jb)
print("Nilai Kritis Chi-Square =", chi_square_critical)
print("Keputusan =", keputusan)
```

```python
#FUNGSI UJI NORMALITAS JARQUE BERA BAGIAN A NOMOR 2
import math

def kolmogorov_smirnov_test(data):
    # Langkah 1: Menghitung jumlah data
    n = len(data)

    # Langkah 2: Mengurutkan data
    data_urut = sorted(data)

    # Langkah 3: Menghitung mean
    mean = sum(data) / n

    # Langkah 4: Menghitung jumlah kuadrat
    jumlah_kuadrat = 0
    for x in data:
        jumlah_kuadrat += (x - mean) ** 2

    # Langkah 5: Menghitung varians
    varians = jumlah_kuadrat / n

    # Langkah 6: Menghitung standar deviasi
    std = varians ** 0.5

    # Langkah 7: Menghitung ECDF
    ecdf = []
    for i in range(n):
        ecdf.append((i + 1) / n)

    # Langkah 8: Menghitung Z-Score
    z_score = []
    for x in data_urut:
        z = (x - mean) / std
        z_score.append(z)

    # Langkah 9: Menghitung CDF Normal Teoritis
    cdf_normal = []
    for z in z_score:
        cdf = 0.5 * (1 + math.erf(z / math.sqrt(2)))
        cdf_normal.append(cdf)

    # Langkah 10: Menghitung selisih |ECDF - CDF|
    selisih = []
    for i in range(n):
        d = abs(ecdf[i] - cdf_normal[i])
        selisih.append(d)

    # Langkah 11: Menghitung statistik K-S
    D_hitung = max(selisih)

    # Langkah 12: Menghitung nilai kritis K-S
    D_tabel = 1.36 / (n ** 0.5)

    # Langkah 13: Membuat keputusan
    if D_hitung > D_tabel:
        keputusan = "Tolak H0: Data tidak berdistribusi normal"
    else:
        keputusan = "Gagal Menolak H0: Data berdistribusi normal"

    # OUTPUT SELURUH HASIL
    print("HASIL UJI KOLMOGOROV-SMIRNOV:")

    print("Data Awal =", data)
    print("Data Terurut =", data_urut)

    print("\nn =", n)

    print("\nMean =", mean)

    print("\nJumlah Kuadrat =", jumlah_kuadrat)

    print("\nVarians =", varians)

    print("\nStandar Deviasi =", std)

    print("\nECDF =")
    print(ecdf)

    print("\nZ-Score =")
    print(z_score)

    print("\nCDF Normal =")
    print(cdf_normal)

    print("\nSelisih =")
    print(selisih)

    print("\nD Hitung =", D_hitung)

    print("\nD Tabel =", D_tabel)

    print("\nKeputusan =", keputusan)

# Menjalankan fungsi
kolmogorov_smirnov_test(data)
```

```python
#Bagian B UJI HOMOGENITAS VARIAN (F Varians)
data1 = [70, 75, 80, 72, 78]
data2 = [65, 85, 60, 90, 70]
```

Pada tahap ini, data dimasukkan ke dalam dua buah list Python yang mewakili dua kelompok yang akan dibandingkan variansnya.

data1 berisi nilai dari kelompok pertama.
data2 berisi nilai dari kelompok kedua.

Data tersebut akan digunakan untuk menghitung:

1. Jumlah data masing-masing kelompok.
2. Rata-rata (mean)masing-masing kelompok.
3. Deviasi absolut terhadap rata-rata kelompok.
4. Statistik uji Levene.

Tujuan pengujian adalah untuk mengetahui apakah kedua kelompok memiliki varians yang sama (homogen) atau berbeda (tidak homogen).

```python
# NOMOR 1B
# Langkah 1: Menghitung jumlah observasi
n1 = len(data1)
n2 = len(data2)

# Langkah 2: Menghitung mean sampel pertama
mean1 = sum(data1) / n1

# Langkah 3: Menghitung mean sampel kedua
mean2 = sum(data2) / n2

# Langkah 4: Menghitung varians sampel pertama
jumlah_kuadrat1 = 0

for x in data1:
    jumlah_kuadrat1 += (x - mean1) ** 2

varians1 = jumlah_kuadrat1 / (n1 - 1)

# Langkah 5: Menghitung varians sampel kedua
jumlah_kuadrat2 = 0

for x in data2:
    jumlah_kuadrat2 += (x - mean2) ** 2

varians2 = jumlah_kuadrat2 / (n2 - 1)

# Langkah 6: Menghitung F hitung
if varians1 > varians2:
    f_hitung = varians1 / varians2
    df1 = n1 - 1
    df2 = n2 - 1
else:
    f_hitung = varians2 / varians1
    df1 = n2 - 1
    df2 = n1 - 1

# Langkah 7: Nilai F tabel
f_tabel = 6.388

# Langkah 8: Keputusan
if f_hitung > f_tabel:
    keputusan = "Tolak H0: varians tidak homogen"
else:
    keputusan = "Gagal Menolak H0: varians homogen"

print("HASIL UJI F VARIANS:")

print("\nJumlah Data")
print("n1 =", n1)
print("n2 =", n2)

print("\nRata-rata")
print("mean1 =", mean1)
print("mean2 =", mean2)

print("\nJumlah Kuadrat")
print("jumlah_kuadrat1 =", jumlah_kuadrat1)
print("jumlah_kuadrat2 =", jumlah_kuadrat2)

print("\nVarians")
print("varians1 =", varians1)
print("varians2 =", varians2)

print("\nDerajat Bebas")
print("df1 =", df1)
print("df2 =", df2)

print("\nStatistik Uji")
print("F Hitung =", f_hitung)
print("F Tabel =", f_tabel)

print("\nKeputusan")
print(keputusan)
```

```python
# Nomor 2B

data1 = [70, 75, 80, 72, 78]
data2 = [65, 85, 60, 90, 70]

# Langkah 1: Menghitung jumlah data
n1 = len(data1)
n2 = len(data2)

# Langkah 2: Menghitung mean masing-masing kelompok
mean1 = sum(data1) / n1
mean2 = sum(data2) / n2

# Langkah 3: Menghitung deviasi absolut
z1 = []
for x in data1:
    z1.append(abs(x - mean1))

z2 = []
for x in data2:
    z2.append(abs(x - mean2))

# Langkah 4: Menghitung mean deviasi absolut
mean_z1 = sum(z1) / n1
mean_z2 = sum(z2) / n2

# Langkah 5: Menghitung grand mean
grand_mean = (sum(z1) + sum(z2)) / (n1 + n2)

# Langkah 6: Menghitung Jumlah Kuadrat Antar Kelompok
jumlah_kuadrat_antar_kelompok = (n1 * (mean_z1 - grand_mean) ** 2+ n2 * (mean_z2 - grand_mean) ** 2)

# Langkah 7: Menghitung Jumlah Kuadrat Dalam Kelompok
jumlah_kuadrat_dalam_kelompok = 0

for z in z1:
    jumlah_kuadrat_dalam_kelompok += (z - mean_z1) ** 2

for z in z2:
    jumlah_kuadrat_dalam_kelompok += (z - mean_z2) ** 2

# Langkah 8: Menghitung statistik Levene
N = n1 + n2
k = 2

levene = ((N - k) / (k - 1)) * (
    jumlah_kuadrat_antar_kelompok /
    jumlah_kuadrat_dalam_kelompok
)

# Langkah 9: Menentukan F tabel
f_tabel = 5.32

# Langkah 10: Keputusan
if levene > f_tabel:
    keputusan = "Tolak H0: Varians tidak homogen"
else:
    keputusan = "Gagal Menolak H0: Varians homogen"

# Output
print("HASIL UJI LEVENE:")

print("\nJumlah Data")
print("n1 =", n1)
print("n2 =", n2)

print("\nMean Kelompok")
print("mean1 =", mean1)
print("mean2 =", mean2)

print("\nDeviasi Absolut")
print("z1 =", z1)
print("z2 =", z2)

print("\nMean Deviasi Absolut")
print("mean_z1 =", mean_z1)
print("mean_z2 =", mean_z2)

print("\nGrand Mean")
print("grand_mean =", grand_mean)

print("\nJumlah Kuadrat")
print("Jumlah Kuadrat Antar Kelompok =", jumlah_kuadrat_antar_kelompok)
print("Jumlah Kuadrat Dalam Kelompok =", jumlah_kuadrat_dalam_kelompok)

print("\nStatistik Uji")
print("Levene Hitung =", levene)
print("F Tabel =", f_tabel)

print("\nKeputusan")
print(keputusan)
```

Pada tahap ini, data dimasukkan ke dalam sebuah list dua dimensi (matriks) bernama data3. Setiap baris merepresentasikan kategori jenis kelamin dan setiap kolom merepresentasikan kategori preferensi minuman.

Data tersebut akan digunakan untuk menghitung:

1. Jumlah baris.
2. Jumlah kolom.
3. Total setiap baris.
4. Total setiap kolom.
5. Grand total.
6. Frekuensi harapan.
7. Nilai Chi-Square hitung.
8. Nilai Chi-Square tabel.
9. Keputusan pengujian.

Uji Chi-Square digunakan untuk mengetahui apakah terdapat hubungan (asosiasi) antara dua variabel kategorik, yaitu jenis kelamin dan preferensi minuman.

```python
# NOMOR 1
# Langkah 1: Menghitung jumlah baris dan kolom
jumlah_baris = len(data3)
jumlah_kolom = len(data3[0])

# Langkah 2: Menghitung total setiap baris
total_baris = []

for baris in data3:
    total_baris.append(sum(baris))

# Langkah 3: Menghitung total setiap kolom
total_kolom = []

for j in range(jumlah_kolom):

    total = 0

    for i in range(jumlah_baris):
        total += data3[i][j]

    total_kolom.append(total)

# Langkah 4: Menghitung grand total
grand_total = sum(total_baris)

# Langkah 5: Menghitung frekuensi harapan
harapan = []
for i in range(jumlah_baris):
    baris = []
    for j in range(jumlah_kolom):
        frekuensi_harapan = (total_baris[i] *
        total_kolom[j]) / grand_total
        baris.append(frekuensi_harapan)
    harapan.append(baris)

# Langkah 6: Menghitung Chi-Square Hitung
chi_square = 0
for i in range(jumlah_baris):
    for j in range(jumlah_kolom):
        frekuensi_observasi = data3[i][j]
        frekuensi_harapan = harapan[i][j]
        chi_square += ((frekuensi_observasi - frekuensi_harapan) ** 2) / frekuensi_harapan

# Langkah 7: Menghitung derajat bebas
derajat_bebas = ((jumlah_baris - 1)*(jumlah_kolom - 1))

# Langkah 8: Menentukan Chi-Square Tabel
chi_tabel = 3.841

# Langkah 9: Keputusan
if chi_square > chi_tabel:
    keputusan = "Tolak H0: Terdapat hubungan"
else:
    keputusan = "Gagal Menolak H0: Tidak terdapat hubungan"

# Output
print("HASIL UJI CHI-SQUARE:")

print("\nJumlah Baris =", jumlah_baris)
print("Jumlah Kolom =", jumlah_kolom)

print("\nTotal Baris =", total_baris)
print("Total Kolom =", total_kolom)

print("\nGrand Total =", grand_total)

print("\nFrekuensi Harapan =")
for baris in harapan:
    print(baris)

print("\nDerajat Bebas =", derajat_bebas)

print("\nChi-Square Hitung =", chi_square)
print("Chi-Square Tabel =", chi_tabel)

print("\nKeputusan =")
print(keputusan)
```

Karena nilai Chi-Square hitung lebih besar daripada nilai Chi-Square tabel, maka H₀ ditolak. Sehingga Terdapat hubungan antara jenis kelamin dan preferensi minuman. Dengan kata lain, kedua variabel tidak saling independen.

```python
# Nomor 2

# Data dari soal
data3 = [[30, 20], [15, 35]]

# Fungsi faktorial
def faktorial(n):
    hasil = 1
    for i in range(1, n + 1):
        hasil *= i
    return hasil

# Langkah 1: Mengambil nilai setiap sel
a = data3[0][0]
b = data3[0][1]
c = data3[1][0]
d = data3[1][1]

# Langkah 2: Menghitung total baris
total_baris_1 = a + b
total_baris_2 = c + d

# Langkah 3: Menghitung total kolom
total_kolom_1 = a + c
total_kolom_2 = b + d

# Langkah 4: Menghitung grand total
grand_total = a + b + c + d

# Langkah 5: Menghitung probabilitas Fisher
probabilitas_fisher = (
    faktorial(total_baris_1)
    * faktorial(total_baris_2)
    * faktorial(total_kolom_1)
    * faktorial(total_kolom_2)
) / (
    faktorial(a)
    * faktorial(b)
    * faktorial(c)
    * faktorial(d)
    * faktorial(grand_total))

# Langkah 6: Menentukan tingkat signifikansi
alpha = 0.05

# Langkah 7: Menentukan keputusan
if probabilitas_fisher < alpha:
    keputusan = "Tolak H0: Terdapat hubungan antara kedua variabel"
else:
    keputusan = "Gagal Menolak H0: Tidak terdapat hubungan antara kedua variabel"

# Output
print("HASIL FISHER'S EXACT TEST:")

print("\nNilai Setiap Sel")
print("a =", a)
print("b =", b)
print("c =", c)
print("d =", d)

print("\nTotal Baris")
print("Total Baris 1 =", total_baris_1)
print("Total Baris 2 =", total_baris_2)

print("\nTotal Kolom")
print("Total Kolom 1 =", total_kolom_1)
print("Total Kolom 2 =", total_kolom_2)

print("\nGrand Total")
print("Grand Total =", grand_total)

print("\nProbabilitas Fisher")
print("Probabilitas Fisher =", probabilitas_fisher)

print("\nTingkat Signifikansi")
print("Alpha =", alpha)

print("\nKeputusan")
print(keputusan)
```
